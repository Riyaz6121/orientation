```python
from shapely import wkb
from shapely.geometry import Point, LineString
from shapely.ops import nearest_points
```

```python
if 'geom' in properties.columns:
    # Convert WKB to geometry objects - remove bytes() wrapper
    properties['geometry'] = properties['geom'].apply(lambda x: wkb.loads(x))
    properties_gdf = gpd.GeoDataFrame(properties, geometry='geometry', crs='EPSG:4283')  # GDA94
else:
    # Create point geometries from lat/lon
    properties_gdf = gpd.GeoDataFrame(
        properties,
        geometry=gpd.points_from_xy(properties.longitude, properties.latitude),
        crs='EPSG:4326'
    )

print("Properties GeoDataFrame created successfully")
print("Sample geometry:", properties_gdf.geometry.iloc[0])
print("CRS:", properties_gdf.crs)
```

```python
# Ensure all datasets are in the same CRS
properties_gdf = properties_gdf.to_crs(cadastre.crs)
roads = roads.to_crs(cadastre.crs)

print("All datasets in CRS:", cadastre.crs)
```

```python
# Step 1: Spatially join properties with cadastre to get property boundaries
properties_with_cadastre = gpd.sjoin(
    properties_gdf, 
    cadastre, 
    how='left', 
    predicate='within'
)
```

```python
properties_with_cadastre.columns.tolist()
```

```python
print(f"\nProperties matched to cadastre: {properties_with_cadastre['sa4'].notna().sum()}")
```

```python
# Step 2: For each property, find the nearest road
def find_nearest_road(property_point, roads_gdf):
    """Find the nearest road to a property"""
    distances = roads_gdf.geometry.distance(property_point)
    nearest_idx = distances.idxmin()
    return roads_gdf.loc[nearest_idx], distances[nearest_idx]

# Step 3: Calculate orientation based on property boundary and nearest road
def calculate_orientation(property_geom, cadastre_geom, nearest_road_geom):
    """
    Calculate the orientation of a property based on:
    - The property boundary from cadastre
    - The nearest road
    
    Logic: Find which edge of the property polygon is closest to the road,
    then calculate the bearing of that edge
    """
    if pd.isna(cadastre_geom) or cadastre_geom is None:
        return None, None
    
    # Handle MultiPolygon - get the largest polygon
    if cadastre_geom.geom_type == 'MultiPolygon':
        cadastre_geom = max(cadastre_geom.geoms, key=lambda p: p.area)
    
    if cadastre_geom.geom_type != 'Polygon':
        return None, None
    
    # Get all edges of the property boundary
    coords = list(cadastre_geom.exterior.coords)
    
    # Find the edge closest to the road
    min_distance = float('inf')
    closest_edge = None
    
    for i in range(len(coords) - 1):
        edge = LineString([coords[i], coords[i + 1]])
        distance = edge.distance(nearest_road_geom)
        
        if distance < min_distance:
            min_distance = distance
            closest_edge = edge
    
    if closest_edge is None:
        return None, None
    
    # Calculate the bearing of the closest edge
    x1, y1 = closest_edge.coords[0]
    x2, y2 = closest_edge.coords[1]
    
    # Calculate angle in degrees (0 = East, 90 = North)
    angle = np.degrees(np.arctan2(y2 - y1, x2 - x1))
    
    # Normalize to 0-360 degrees
    angle = (angle + 360) % 360
    
    # Convert to compass bearing (0 = North, 90 = East)
    bearing = (90 - angle) % 360
    
    return bearing, min_distance

def bearing_to_direction(bearing):
    """Convert bearing to cardinal direction"""
    if bearing is None:
        return 'Unknown'
    
    directions = ['North', 'North-East', 'East', 'South-East', 
                  'South', 'South-West', 'West', 'North-West']
    
    # Divide 360 degrees into 8 segments (45 degrees each)
    # Add 22.5 to center each direction in its segment
    index = int((bearing + 22.5) / 45) % 8
    
    return directions[index]
```

```python
# Process each property
results = []

print("\nProcessing properties...")
for idx, row in properties_with_cadastre.iterrows():
    # Get property location
    property_point = row.geometry
    
    # Find nearest road
    nearest_road, road_distance = find_nearest_road(property_point, roads)
    
    # Get cadastre geometry for this property
    cadastre_geom = None
    if pd.notna(row.get('sad4')):
        cadastre_match = cadastre[cadastre['sad4'] == row['sad4']]
        if not cadastre_match.empty:
            cadastre_geom = cadastre_match.iloc[0].geometry
    
    # Calculate orientation
    bearing, edge_distance = calculate_orientation(
        property_point, 
        cadastre_geom, 
        nearest_road.geometry
    )
    
    direction = bearing_to_direction(bearing)
    
    results.append({
        'address': row.get('address', 'N/A'),
        'orientation': direction,
        'bearing': bearing,
        'nearest_road': nearest_road.get('name', 'Unnamed'),
        'road_distance_m': round(road_distance, 2) if not np.isnan(road_distance) else None
    })
    
    # Progress indicator
    if idx % 1000 == 0:
        print(f"Processed {idx} properties...")
```

```python
# Create output dataframe
output_df = pd.DataFrame(results)
```

```python
# Remove duplicates
output_df = output_df.drop_duplicates(subset=['address'])
```

```python
print(f"\n=== RESULTS ===")
print(f"Total properties: {len(output_df)}")
print(f"\nOrientation distribution:")
print(output_df['orientation'].value_counts())

print(f"\nSample results:")
print(output_df.head(10))

# Save to file
output_df.to_csv('property_orientations.csv', index=False)
print(f"\nResults saved to 'property_orientations.csv'")
```
